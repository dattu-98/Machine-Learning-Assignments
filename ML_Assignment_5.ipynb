**2. Describe quantitative and qualitative data in depth. Make a distinction between the two.**

1. Quantitative Data: Quantitative data are measures of values or counts and are expressed as numbers. This can be answered like 'How Many' or 'How much'
> Ex: Price of a product, no. of products sold, profit, volume of the traded quantity etc...

2. Qualitative Data: Qualitative data is data that is not easily reduced to numbers. Qualitative data tends to answer questions about the 'what', 'how' and 'why' of a phenomenon, rather than questions of 'how many' or 'how much'
> Ex: Difficulty level of the problem, Size of a product, Taste of a dish etc...

2. Imputing the missing values with the mean, median of that particular column.

**8. What are the various data pre-processing techniques? Explain dimensionality reduction and function selection in a few words.**

Data preprocessing is widely classified into three segments:
1. Data Cleaning: Handling the missing data by Ignoring the Tuples, Filling the missing values, Handling of Noisy data
2. Data Transformation: Normalization of Data, Selection of Input attributes
3. Data Reduction: Dimensionality reduction, Attribute Subset Selection

![](https://media.geeksforgeeks.org/wp-content/uploads/20190312184006/Data-Preprocessing.png)


Dimensionality Reduction: It is a process of where variables are transformed to a lower dimension. This reduce the size of data by encoding mechanisms.It can be lossy or lossless. If after reconstruction from compressed data, original data can be retrieved, such reduction are called lossless reduction else it is called lossy reduction. The two effective methods of dimensionality reduction are:Wavelet transforms and PCA (Principal Component Analysis)

1. IQR is called as Inter Quantile Range. The IQR describes the middle 50% of values when ordered from lowest to highest. To find the interquartile range (IQR), first find the median (middle value) of the lower and upper half of the data. These values are quartile 1 (Q1) and quartile 3 (Q3). The IQR is the difference between Q3 and Q1.
